<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
# @@@ 5
df['REVENUE_PER_STUDENT'] = df['TOTAL_REVENUE'] / df['ENROLL_ALL_EST']
df['INSTRUCTION_EXP_RATIO'] = df['INSTRUCTION_EXPENDITURE'] / df['TOTAL_EXPENDITURE']
df['SUPPORT_SERVICES_RATIO'] = df['SUPPORT_SERVICES_EXPENDITURE'] / df['TOTAL_EXPENDITURE']
```

Feature engineering justification:
1. States with higher revenue per student might have more resources for education, potentially improving student performance.
2. Higher investment in instruction may correlate with better test scores.
3. Support services might indirectly improve performance by addressing non-academic needs.

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
# @@@ 8

features = ['REVENUE_PER_STUDENT', 'INSTRUCTION_EXP_RATIO', 'SUPPORT_SERVICES_RATIO']
X = df[features].fillna(0)
y = df['AVG_MATH_4_SCORE']
```

```python
# @@@ 9 

# 70/30 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# @@@ 10
# import your sklearn class here

from sklearn.linear_model import LinearRegression
```

```python
# @@@ 11
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
# @@@ 12

r_squared=model.score(X_test, y_test)
print('R-squared:', r_squared)

mean_error=np.mean(model.predict(X_test)-y_test)
print('Mean Error:', mean_error)

mean_absolute_error=np.mean(np.abs(model.predict(X_test)-y_test))
print('Mean Absolute Error:', mean_absolute_error)

rmse=np.mean((model.predict(X_test)-y_test)**2)**0.5
print('Root Mean Squared Error:', rmse)
```
