```python
# You may include any package you deem fit. We sugggest looking into Scikit-learn
import pandas as pd
import seaborn as sns
import numpy as np
```

## Dataset


```python
# DO NOT MODIFY
loan_data = pd.read_csv("../../data/loan_approval.csv")
```

## EDA
Uncomment to see desired output. Add more analysis if you like

```python

# import matplotlib.pyplot as plt

# # ------ Display basic information ------
print(loan_data.columns)
print(loan_data.describe())

# # ------ Check for missing values ------
print(loan_data.isnull().sum())

# # ------ Visualize the distribution of loan status ------
loan_status_counts = loan_data['loan_status'].value_counts()
plt.bar(loan_status_counts.index, loan_status_counts.values)
plt.title('Distribution of Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('Count')

# # ------ Visualize the distribution of numerical features ------ 
loan_data.hist(bins=30, figsize=(20, 15))

# # ------ Correlation matrix ------
# # Encoding
loan_data['education'] = loan_data['education'].map({'Graduate': 1, 'Not Graduate': 0})
loan_data['self_employed'] = loan_data['self_employed'].map({'Yes': 1, 'No': 0})
loan_data['loan_status'] = loan_data['loan_status'].map({'Approved': 1, 'Rejected': 0})

corr_matrix = loan_data.corr()
fig, ax = plt.subplots(figsize=(10, 8))
cax = ax.matshow(corr_matrix, cmap='coolwarm')
fig.colorbar(cax)
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)

# # ----- MORE (Encouraged but not required) ------

# #Categorical var
categorical_cols = ['education', 'self_employed']
for col in categorical_cols:
    print(f"Value counts for {col}:")
    print(loan_data[col].value_counts())
    print()

# #Boxplot numerical
numerical_cols = ['income_annum', 'loan_amount', 'loan_term', 'cibil_score']
for col in numerical_cols:
    sns.boxplot(data=loan_data, x='loan_status', y=col)
    plt.title(f'{col} vs Loan Status')
    plt.show()

# # Countplots categorical
for col in categorical_cols[:-1]:  # Exclude 'loan_status' itself
    sns.countplot(data=loan_data, x=col, hue='loan_status')
    plt.title(f'{col} by Loan Status')
    plt.show()

# #High correlation features
loan_data['loan_status_encoded'] = loan_data['loan_status'].map({'Approved': 1, 'Denied': 0})
correlations_with_target = loan_data.corr()['loan_status_encoded'].sort_values(ascending=False)
print(correlations_with_target)

sns.pairplot(loan_data, vars=['cibil_score', 'income_annum', 'loan_amount'], hue='loan_status')
plt.show()
```

## Feature Engineering

You may want to convert categorical variables to numerical. For example, education takes on the value Graduate and Not Graduate. But we want it to be 0 or 1 for machine learning algorithms to use.

```python
loan_data['debt_to_income_ratio'] = loan_data['loan_amount'] / loan_data['income_annum']
loan_data['assets_total_value'] = loan_data['residential_assets_value'] + loan_data['commercial_assets_value'] + loan_data['luxury_assets_value'] + loan_data['bank_asset_value']

loan_data.head()
```

## Model Selection

You are free to use any classification machine learning models you like: Logistic Regression, Decision Trees/Random Forests, Support Vector Machines, KNN ... 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

features = ['cibil_score', 'loan_amount', 'income_annum', 'debt_to_income_ratio', 'assets_total_value']
X = loan_data[features]
y = loan_data['loan_status']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Logistic Regression model
model = LogisticRegression(max_iter=500)
```

## Model Training and Evaluation

```python
#Training
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Evaluating
print('Accuracy (R2):', model.score(X_test, y_test))

mean_error = np.mean(model.predict(X_test) - y_test)
print('Mean Error:', mean_error)

mean_absolute_error = np.mean(np.abs(model.predict(X_test) - y_test))
print('Mean Absolute Error:', mean_absolute_error)

rmse = np.mean((model.predict(X_test) - y_test)**2)**0.5
print('Root Mean Squared Error (RMSE):', rmse)
```

## Model Optimization and Testing

```python
col_name = 'cibil_score'

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Generate predictions for the test set
y_pred = model.predict(X_test)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()

from sklearn.metrics import classification_report

# Classification report
print('Classification Report:')
print(classification_report(y_test, y_pred))
```
